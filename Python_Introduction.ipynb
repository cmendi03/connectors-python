```python
import time
current_year = int(time.strftime('%Y'))
received_year = int(input("What year did you get your passport?"))

if received_year + 10 < current_year:
    print('You should go get a new passport')
else:
    print('You may use your current passport')
```

- Line `1` is letting the computer know that we are going to need to get some information about time
- Line `2` asks the computer what the current year is and saves it by the name current_year
- Line `3` asks whoever is using this program to enter when he or she got the passport
- Line `4` checks if the passport is invalid, and if so alerts the user of this (line `5`) and if not it uses the alternate case (line `6`) and tells the user the passport is still acceptable (line `7`)

```python
#improved version

def get_year():
    response = input("What year did you get your passport?\n(Numbers only please)")
    try:
        response = int(response)
        if len(str(response)) > 4:
            print("ERROR: Invalid year! Please try again.")
            return get_year()
        elif len(str(response)) < 4:
            print("ERROR: Invalid year! Please try again.")
            return get_year()
        elif response + 10 < current_year:
            return('You should go get a new passport')
        else:
            return('You may use your current passport')
    except:
        print("ERROR: Invalid year! Please try again.")
        return get_year()

get_year()
```

PEMDAS is an acronym that can be used to help remember the precedence / order of operations for algebraic formulas such as these. **PEMDAS stands for "Parentheses, Exponent, Multiplication, Division, Addition, Subtraction**". In the absence of parentheses, you can use this ordering to determine what calculations to do before others. In math, the order of operations matters, and ordering it wrong (for example, just going left-to-right) can lead to a wrong answer! Technically, in python, multiplication and division are "tied" in PEMDAS, and so are addition and subtraction. You can use this slightly modifier ordering:

`(1)` Parentheses <br>
`(2)` Exponents next <br>
`(3)` Multiplication, Division, and Modulus (%) are tied <br>
`(4)` Addition and Subtraction are tied at the bottom

`(3)` Next up would be the 40 / 2 because it takes higher precedence than the 2 + 250. 40 / 2 Would yield 20.0. <br>

## Dates

Representing dates and times in programming is a challenge. First of all, we have to represent them in a standard, universally accepted format. Fortunately, the `International Organization for Standardization` **(ISO)** developed a worldwide standard `ISO 8601`, which represents date- and time-related objects as `YYYY-MM-DD HH:MM:SS` with information ranging from the most important (years,YYYY) to the least important (seconds, SS). Each piece of this format is represented as a four- or two-digit number.

**The datetime module in Python has 5 main classes** (parts of the module):

- `date` to manipulate date objects
- `time` to manipulate time objects
- `datetime` is a combination of date and time
- `timedelta` allows us to work with time duration
- `tzinfo` allows us to work with timezones

Additionally, we'll use the `zoneinfo` module, which provides us a modern way of working with time zones, and the `dateutil` package, which contains plenty of useful functions to work with dates and times.

Let's import the datetime module and create our first date and time objects:

```python
# From the datetime module import date
from datetime import date

# Create a date object of 2000-02-03
date(2000, 2, 3)
```

#### Extracting dates from strings

In `data science`, and in programming generally, we mostly work with dates and times that are stored as strings in dozens of different formats, depending on the region, company, or the granularity of the information we need. Sometimes, we need the date and the exact time, but in other cases, we only need the year and the month. **How can we extract the data we need from a string to easily manipulate it as a datetime (date, time) object?**

**fromisoformat()**, **isoformat()**

The first function we'll learn to convert a date string into a date object is `fromisoformat`. We call it this way because it uses the ISO 8601 format (i.e., YYYY-MM-DD). Let's look at an example:

```python
from datetime import date

# Convert a date string into a date object
date.fromisoformat("2022-12-31")
```

```python
date = date.fromisoformat("2022-12-31")
type(date)
```

We can also perform an `inverse operation` and convert a datetime object into a date string in the ISO format. For this purpose, we should use the `isoformat()`:

```python
from datetime import date

# Convert a datetime object into a string in the ISO format
date_str = date(2022, 12, 31).isoformat()
date_str
```

```python
# Data Type
type(date_str)
```

**strptime()**

We can use the `strptime()` function to convert an arbitrary date/time string into a datetime object. Our string doesn't necessarily need to follow the ISO format, but we should specify which part of the string represents which date or time unit (year, hour, etc.). Let's look at an example to clarify. First of all, we'll use the strict ISO format to convert a string into a datetime object:

```python
from datetime import datetime

# Date as a string
iso_datetime = "2022-12-31 23:59:58"
type(iso_datetime)
```

```python
# ISO format
iso_format = "%Y-%m-%d %H:%M:%S"

# Convert the string into a datetime object
datetime.strptime(iso_datetime, iso_format)
```

```python
type(iso_datetime)
```

- In the first line, we create a date/time string.
- In the second line, we specify the format of the string using a special code that contains a percent sign followed by a character that encodes a date or time unit.
- Finally, in the third line, we use the `strptime()` function to convert the string into a datetime object. This function takes two arguments: the string and the format of the string.

```python
from datetime import datetime

# Date as a string
iso_datetime = "2022-12-31 23:59:58.029796"

# ISO format
iso_format = "%Y-%m-%d %H:%M:%S.%f"

# Convert the string into a datetime object
datetime.strptime(iso_datetime, iso_format)
```

#### Converting datetime objects into strings

**strftime()**

In Python, we can also convert a datetime object into a string using the `strftime()` function. It takes two arguments: a datetime object and the format of the output string.

```python
# Create a datetime object
datetime_obj = datetime(2022, 12, 31)

# American date format
american_format = "%m-%d-%Y"

# European format
european_format = "%d-%m-%Y"
```

```python
datetime.strftime(datetime_obj, american_format)
```

#### Arithmetic operations with Dates

Sometimes, we may want to compute the difference between two dates or perform other arithmetic operations on dates and times. Fortunately, Python has many instruments in its toolkit to perform such calculations.

The first operation we can perform is to calculate the `difference between two dates`. To do this, we use the minus sign:

```python
from datetime import datetime

# Instantiate two dates
first_date = datetime(2022, 1, 1)
second_date = datetime(2022, 12, 31)

# Difference between two dates
date_diff = second_date - first_date

print("The number of days and hours between " + datetime.strftime(first_date, "%Y-%m-%d") + " and " +
      datetime.strftime(second_date, "%Y-%m-%d") + " is " + str(date_diff) + ".")
```

```python
date_diff
```

```python
date_diff
```

```python
type(date_diff)
```

The type of this object is `datetime.timedelta`. It has delta in its name, which refers to a Greek letter delta, $\Delta$, which in science and engineering, describes a change. Indeed, here it represents a change (difference) in time.

<br>
What if we're interested only in the number of days between two dates? We can:

```python
print(f'The number of days between {datetime.strftime(first_date, "%Y-%m-%d")} and {datetime.strftime(second_date, "%Y-%m-%d")} is {(date_diff).days}.')
```
